```python
# Gerekli kütüphaneleri içe aktar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# Veri setini yükle
dataset = pd.read_csv("/dataset.csv")  # Dosya adını ve yolunu burada belirtin
dataset.tail()
```

```python
# Bağımsız (X) ve bağımlı (y) değişkenleri ayır
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Veriyi eğitim ve test setlerine ayır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Doğrusal regresyon modelini oluştur ve eğit
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python

# Test seti üzerinde tahmin yap
y_pred = regressor.predict(X_test)
```

```python
# Görselleştirme: Eğitim Verileri ve Model
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
plt.plot(X_train, regressor.predict(X_train), color='blue')  # Modelin tahmin ettiği doğrusal çizgiyi maviyle göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli")
plt.xlabel("Kıdem")
plt.ylabel("Maaş")
plt.show()

# Görselleştirme: Eğitim Verileri ve Tahmin
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
modelin_tahmin_ettigi_y = regressor.predict(X_train)
plt.scatter(X_train, modelin_tahmin_ettigi_y, color='blue')  # Tahminleri mavi noktalarla göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli")
plt.xlabel("Kıdem")
plt.ylabel("Maaş")
plt.show()

# Görselleştirme: Test Verileri ve Tahmin
plt.scatter(X_test, y_pred, color='green')  # Test verilerinin tahminlerini yeşil noktalarla göster
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli (Test)")
plt.xlabel("Kıdem")
plt.ylabel("Maaş")
plt.show()
```

```python
# Görselleştirme: Test Verileri ve Tahmin
plt.scatter(X_test, y_pred, color='green')  # Test verilerinin tahminlerini yeşil noktalarla göster
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli (Test)")
plt.xlabel("year")
plt.ylabel("cited")
plt.show()
```

```python
# Görselleştirme: Eğitim Verileri ve Tahmin
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
modelin_tahmin_ettigi_y = regressor.predict(X_train)
plt.scatter(X_train, modelin_tahmin_ettigi_y, color='blue')  # Tahminleri mavi noktalarla göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli")
plt.xlabel("cited")
plt.ylabel("year")
plt.show()
```

```python
# Görselleştirme: Eğitim Verileri ve Model
plt.scatter(X_train, y_train, color='red')  # Eğitim verilerini kırmızı noktalarla göster
plt.plot(X_train, regressor.predict(X_train), color='blue')  # Modelin tahmin ettiği doğrusal çizgiyi maviyle göster
plt.title("Kıdeme Göre Maaş Tahmini Regresyon Modeli")
plt.xlabel("cited")
plt.ylabel("year")
plt.show()
```

```python
# Metrikler
print("R kare skoru: ", r2_score(y_test, y_pred))
print("Ortalama Mutlak Hata: {} \nOrtalama Karesel Hata: {}".format(
    mean_absolute_error(y_test, y_pred),
    mean_squared_error(y_test, y_pred)
))
```
